```python
results = eval(open('pdb_metal_validation_results.txt').read())
```

```python
results
```

```python
results2 = []

for i in results:
    for j in i:
        results2.append(j)
```

```python
len(results2)
```

```python
results3 = []

for i in results2:
    if i[0] == i[2]:
        results3.append(i)
```

```python
len(results3)
```

```python
results3
```

```python
results3_res = []

for i in results3:
    results3_res.append(i[0])
```

```python
set(results3_res)
```

```python
from collections import Counter
```

```python
Counter(results3_res)
```

```python
results3_res_dict = dict(Counter(results3_res))
```

```python
total = 0

for i in results3_res_dict:
    total += results3_res_dict[i]
```

```python
from __future__ import division
```

```python
results3_res_dict_perc = {}

for i in results3_res_dict:
    results3_res_dict_perc[i] = results3_res_dict[i] / total * 100
```

```python
sorted(results3_res_dict_perc.items(), key = lambda x: x[1], reverse = True)
```
