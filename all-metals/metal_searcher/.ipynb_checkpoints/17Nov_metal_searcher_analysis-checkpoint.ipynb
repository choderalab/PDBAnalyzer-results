```python
metals_list = ['TH',
 'BA',
 'YB',
 'EU',
 'FE',
 'DY',
 'V',
 'GA',
 'HG',
 'PR',
 'NI',
 'PT',
 'LA',
 'NA',
 'LI',
 'PB',
 'RE',
 'TL',
 'LU',
 'RU',
 'RB',
 'TE',
 'TB',
 'K',
 'ZN',
 'CO',
 'PD',
 'AG',
 'CA',
 'IR',
 'AM',
 'AL',
 'CE',
 'CD',
 'GD',
 'AU',
 'W',
 'IN',
 'CS',
 'CR',
 'CU',
 'MG',
 'SR',
 'MO',
 'MN',
 'SM',
 'SB',
 'OS',
 'HO']
```

```python
residues_metals = metals_list
```

```python
counter = {}

for i in residues_metals:
    counter[i] = 0
```

```python
for i in residues:
    for j in counter:
        if i == j:
            counter[j] += 1
```

```python
counter
```

```python
from __future__ import division
```

```python
counter_perc = {}

total = 0

for i in counter:
    total += counter[i]
    
for i in counter:
    counter_perc[i] = counter[i] / total * 100
    
counter_perc    
```

```python
sorted_perc = sorted(counter_perc.items(), key = lambda x: x[1], reverse = True)
```

```python
sorted_perc
```

```python
sorted_perc_big = []

for i in sorted_perc:
    if i[1] > 1:
        sorted_perc_big.append(i)
```

```python
sorted_perc_big
```

```python
total_big = 0

for i in sorted_perc_big:
    total_big += i[1]
    
total_big    
```

```python
sorted_perc_big.append(('Others', 100 - total_big))
```

```python
sorted_perc_big
```

```python
sorted_perc_big_corr = [(), (), (), (), (), (), (), (), (), (), (), (), ()]
sorted_perc_big_corr[0] = ('Mg', sorted_perc_big[0][1])
sorted_perc_big_corr[1] = ('Zn', sorted_perc_big[1][1])
sorted_perc_big_corr[2] = ('Ca', sorted_perc_big[2][1])
sorted_perc_big_corr[3] = ('Na', sorted_perc_big[3][1])
sorted_perc_big_corr[4] = ('Mn', sorted_perc_big[4][1])
sorted_perc_big_corr[5] = ('K', sorted_perc_big[5][1])
sorted_perc_big_corr[6] = ('Cd', sorted_perc_big[6][1])
sorted_perc_big_corr[7] = ('Fe', sorted_perc_big[7][1])
sorted_perc_big_corr[8] = ('Sr', sorted_perc_big[8][1])
sorted_perc_big_corr[9] = ('Cu', sorted_perc_big[9][1])
sorted_perc_big_corr[10] = ('Ni', sorted_perc_big[10][1])
sorted_perc_big_corr[11] = ('Co', sorted_perc_big[11][1])
sorted_perc_big_corr[12] = ('Others', sorted_perc_big[12][1])
```

```python
perc_others = []

for i in sorted_perc:
    if i[1] <= 1:
        perc_others.append(i)
```

```python
perc_others
```

```python
others_total = 0

for i in perc_others:
    others_total += i[1]
    
others_total, 100 - others_total
```

```python
100 / others_total
```

```python
perc_others_mult = []

for i in perc_others:
    perc_others_mult.append((i[0], i[1] * (100/others_total)))
    
perc_others_mult    
```

```python
new_total = 0

for i in perc_others_mult:
    new_total += i[1]
```
