```python
analysis = results['over_four_analysis']
```

```python
analysis
```

```python
analysis2 = []

for i in analysis:
    if i != [[], []]:
        analysis2.append(i)
```

```python
analysis2
```

```python
analysis3 = []

for i in analysis2:
    new_list = [[], 0, 0]
    new_list[0] = [x for x in i[0] if x not in ['HIS', 'CYS', 'GLU', 'ASP']]
    new_list[1] = len(new_list[0])
    new_list[2] = len(i[0])
    analysis3.append(new_list)
    
    
    
```

```python
distances = []

for i in analysis2:
    distances.append(len(i[0]))
```

```python
from collections import Counter
```

```python
Counter(distances)
```

```python
analysis3
```

```python
some_numbers = []

for i in analysis3:
    some_numbers.append(i[2] - i[1])
```

```python
Counter(some_numbers)
```

```python
analysis4 = []

for i in analysis2:
    to_append = [[], [], []]
    for j in i[0]:
        if j not in ['HIS', 'CYS', 'ASP', 'GLU']:
            to_append[0].append(j)
            to_append[1].append(i[1][i[0].index(j)])
    to_append[2] = i[1]
    analysis4.append(to_append)                            
                                

```

```python
analysis4
```

```python
len(analysis4)
```

```python
len([i for i in analysis4 if i[0]])
```

```python
analysis5 = [i for i in analysis4 if i[0]]
```

```python
is_true = []

for i in analysis5:
    is_true_is_true = True
    for j in [x for x in i[2] if x not in i[1]]:
        if j > max(i[1]):
            is_true_is_true = False
    is_true.append(is_true_is_true)       
            
```

```python
Counter(is_true)
```
