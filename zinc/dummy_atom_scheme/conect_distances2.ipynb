```python
from numpy import *
```

```python
results = eval(open('24Nov_dummy_atom_scheme_conect_distances.txt').read())
```

```python
results
```

```python
ligand_distances = results['ligand_distances']
```

```python
all_distances = []

for i in ligand_distances:
    for j in i:
        all_distances.append(j)
```

```python
all_distances
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.hist(all_distances, bins = 5, range = (0, 0.5))
```
