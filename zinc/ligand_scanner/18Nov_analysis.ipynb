```python
results_dict = eval(results.read())
```

```python
conect_and_extra_ligand_numbers = results_dict['CONECT_and_extra_ligand_numbers']
```

```python
conect_only_ligand_numbers = results_dict['CONECT_only_ligand_numbers']
```

```python
from collections import Counter
```

```python
conect_only_ligand_numbers_counter = dict(Counter(conect_only_ligand_numbers))
```

```python
conect_only_ligand_numbers_counter
```

```python
from __future__ import division
```

```python
conect_only_perc = {}
total = 0

for i in conect_only_ligand_numbers_counter:
    total += conect_only_ligand_numbers_counter[i]
    
for i in conect_only_ligand_numbers_counter:
    conect_only_perc[i] = conect_only_ligand_numbers_counter[i] / total * 100
    
```

```python
conect_only_perc
```

```python
conect_only_perc_items = conect_only_perc.items()
conect_only_perc_items
```

```python
import matplotlib
matplotlib.use('Agg')
from matplotlib import pyplot as plt
```

```python
plt.figure()
plt.bar(range(len(conect_only_perc_items)), [i[1] for i in conect_only_perc_items], align='center')
plt.xticks(range(len(conect_only_perc_items)), [i[0] for i in conect_only_perc_items])
plt.xlabel("Number of CONECT / CUTOFF ligands")
plt.ylabel("%")
plt.title("CONECT = CUTOFF ligand numbers")
plt.ylim(0, 65)
#plt.autoscale()
plt.savefig('CONECT_equal_CUTOFF_ligand_numbers.png')
```

```python
conect_and_extra_ligand_numbers
```

```python
conect_and_extra_counter = dict(Counter(conect_and_extra_ligand_numbers))
```

```python
conect_and_extra_counter
```

```python
conect_and_extra_perc = {}

total = 0

for i in conect_and_extra_counter:
    total += conect_and_extra_counter[i]
    
for i in conect_and_extra_counter:
    conect_and_extra_perc[i] = conect_and_extra_counter[i] / total * 100
```

```python
conect_and_extra_perc
```

```python
conect_and_extra_items = conect_and_extra_perc.items()
```

```python
plt.figure()
plt.bar(range(len(conect_and_extra_items)), [i[1] for i in conect_and_extra_items], align='center')
plt.xticks(range(len(conect_and_extra_items)), [i[0] for i in conect_and_extra_items])
plt.xlabel("Number of CONECT + extra ligands")
plt.ylabel("%")
plt.title("CUTOFF = CONECT + extra ligand numbers")
#plt.ylim(0, 65)
#plt.autoscale()
plt.savefig('CONECT_plus_extra_ligand_numbers.png')
```
