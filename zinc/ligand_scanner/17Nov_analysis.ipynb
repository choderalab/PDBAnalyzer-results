```python
a_r_all_perc = {}

total = 0

for i in atoms_residues_all_counter:
    total += atoms_residues_all_counter[i]
    
for i in atoms_residues_all_counter: 
    a_r_all_perc[i] = atoms_residues_all_counter[i] / total * 100
```

```python
a_r_all_perc
```

```python
sorted(a_r_all_perc.items(), key = lambda x: x[1])
```

```python
a_r_all_perc_big = []

for i in a_r_all_perc.items():
    if i[1] > 5:
        a_r_all_perc_big.append(i)
```

```python
a_r_all_perc_big
```

```python
total_big = 0

for i in a_r_all_perc_big:
    total_big += i[1]
    
total_big    
```

```python
a_r_all_perc_big = sorted(a_r_all_perc_big, key = lambda x: x[1], reverse = True)
```

```python
a_r_all_perc_big
```

```python
a_r_all_perc_big[8] = ('ASP-OD1', a_r_all_perc_big[8][1])
```

```python
total_perc_big = 0
for i in a_r_all_perc_big:
    total_perc_big += i[1]

a_r_all_perc_big.append(('Others', 100 - total_perc_big))
```

```python
a_r_all_perc_big
```

```python
plt.figure(figsize = (12,8))
plt.bar(range(len(a_r_all_perc_big)), [i[1] for i in a_r_all_perc_big], align='center')
plt.xticks(range(len(a_r_all_perc_big)), [i[0] for i in a_r_all_perc_big])
plt.xlabel("Residue, atom")
plt.ylabel("% all extra ligand atoms")
plt.title("CUTOFF = CONECT + extra, extra residues-atoms")
#plt.autoscale()
#plt.ylim(0, 80)
plt.xlim(-0.5, 9.5)
plt.savefig('extra_in_CUTOFF_residues_atoms_fig.png')
```
