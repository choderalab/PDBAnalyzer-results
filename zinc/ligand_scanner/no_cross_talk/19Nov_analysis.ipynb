```python
results = open('ligand_scanner_results_no_cross_talk_2.txt')
```

```python
results_dict = eval(results.read())
```

```python
results_dict
```

```python
cutoff_atoms = results_dict['CUTOFF_atoms']
```

```python
cutoff_atoms_dict = {}

for i in cutoff_atoms:
    cutoff_atoms_dict[(i[0], i[1])] = {}
    
len(cutoff_atoms_dict)    
```

```python
conect_atoms = results_dict['CONECT_atoms']
```

```python
conect_atoms
```

```python
len(conect_atoms)
```

```python
atoms = []

for i in conect_atoms:
    atoms.append((i[2], i[3]))
```

```python
atoms
```

```python
from collections import Counter
```

```python
atoms_counter = dict(Counter(atoms))
```

```python
atoms_counter
```

```python
sorted(atoms_counter.items(), key = lambda x: x[1], reverse = True)
```

```python
from __future__ import division
```

```python
atoms_counter_perc = {}
total = 0

for i in atoms_counter:
    total += atoms_counter[i]
    
for i in atoms_counter:
    atoms_counter_perc[i] = atoms_counter[i] / total * 100
    
```

```python
atoms_counter_perc
```

```python
atoms_counter_perc_sort = sorted(atoms_counter_perc.items(), key = lambda x: x[1], reverse = True)
```

```python
atoms_counter_perc_sort
```

```python
pie_chart_input = []

for i in atoms_counter_perc_sort:
    if i[1] > 0.1:
        pie_chart_input.append(i)
        

pie_chart_input        
        
```

```python
total = 0

for i in pie_chart_input:
    total += i[1]
    
total    
```

```python
pie_chart_input.append(('Others', 100 - total))
```

```python
len(pie_chart_input)
```
