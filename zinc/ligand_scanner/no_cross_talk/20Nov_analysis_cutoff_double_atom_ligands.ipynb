```python
cutoff_atoms_double_ligand_permutation_numbers = {}

for i in cutoff_atoms_dict_double:
    pairs = 0
    combinations = list(itertools.combinations(cutoff_atoms_dict_double[i], 2))
    for j in combinations:
        if j[0][:(j[0].find("-"))] == j[1][:(j[1].find("-"))]:
            pairs += 1
    cutoff_atoms_double_ligand_permutation_numbers[i] = pairs        
```

```python
cutoff_atoms_double_ligand_permutation_numbers
```

```python
for i in cutoff_atoms_double_ligands:
    cutoff_atoms_double_ligands[i].append(cutoff_atoms_double_ligands[i][0] - cutoff_atoms_double_ligand_permutation_numbers[i])
```

```python
# remove the doubling Asp and Glu
cutoff_atoms_dict_double_corr = {}
for i in cutoff_atoms_dict_double:
    
    cutoff_atoms_dict_double_corr[i] = cutoff_atoms_dict_double[i]
    glu_asp = [x for x in cutoff_atoms_dict_double_corr[i] if x[0:3] == 'GLU' or x[0:3] == 'ASP']
    to_remove = []
    
    for j in glu_asp:
        for k in glu_asp:
            if (j != k and j[:(j.find("-"))] == k[:(k.find("-"))] and j not in to_remove and k not in to_remove):
                to_remove.append(k)
                
    for x in to_remove:
        cutoff_atoms_dict_double_corr[i].remove(x)
                
                
                
```
