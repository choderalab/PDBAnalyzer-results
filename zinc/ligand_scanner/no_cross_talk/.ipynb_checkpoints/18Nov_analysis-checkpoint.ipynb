```python
results = open('ligand_scanner_results_no_cross_talk.txt')
```

```python
results_dict = eval(results.read())
```

```python
results_dict
```

```python
ligand_numbers = results_dict['ligand_numbers']
```

```python
ligand_numbers
```

```python
len(ligand_numbers)
```

```python
equal = 0
conect_higher = 0
cutoff_higher = 0

for i in ligand_numbers:
    if i[0] == i[1]:
        equal += 1
    elif i[0] > i[1]:
        conect_higher += 1
    elif i[0] < i[1]:
        cutoff_higher += 1
```

```python
equal
```

```python
conect_higher
```

```python
cutoff_higher
```

```python
ligand_numbers_cutoff = []

for i in ligand_numbers:
    ligand_numbers_cutoff.append(i[1])
```

```python
from collections import Counter
```

```python
cutoff_counter = dict(Counter(ligand_numbers_cutoff))
```

```python
cutoff_counter
```

```python
from __future__ import division
```

```python
cutoff_counter_perc = {}

total = 0

for i in cutoff_counter:
    total += cutoff_counter[i]
    
for i in cutoff_counter:
    cutoff_counter_perc[i] = cutoff_counter[i] / total * 100
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
```

```python
cutoff_counter_items = cutoff_counter_perc.items()
```

```python
plt.figure()
plt.bar(range(len(cutoff_counter_items)), [i[1] for i in cutoff_counter_items], align='center')
plt.xticks(range(len(cutoff_counter_items)), [i[0] for i in cutoff_counter_items])
plt.xlabel("Number of CUTOFF ligands")
plt.ylabel("%")
plt.title("CUTOFF ligand numbers")
#plt.ylim(0, 65)
plt.autoscale()
plt.savefig('CUTOFF_ligand_numbers.png')
```

```python
ligand_numbers_conect = []

for i in ligand_numbers:
    ligand_numbers_conect.append(i[0])
```

```python
conect_counter = dict(Counter(ligand_numbers_conect))
```

```python
conect_counter
```

```python
conect_counter_perc = {}

total = 0 

for i in conect_counter:
    total += conect_counter[i]
    
for i in conect_counter:
    conect_counter_perc[i] = conect_counter[i] / total * 100
```

```python
conect_counter_perc
```

```python
cutoff_minus_conect_perc = {}

for i in cutoff_counter_perc:
    if i != 10:
        cutoff_minus_conect_perc[i] = cutoff_counter_perc[i] - conect_counter_perc[i]
    if i == 10:
        cutoff_minus_conect_perc[i] = cutoff_counter_perc[i]
```

```python
cutoff_minus_conect_perc
```

```python
cutoff_minus_conect_perc_items = cutoff_minus_conect_perc.items()
```

```python
plt.figure()
plt.bar(range(len(cutoff_minus_conect_perc)), [i[1] for i in cutoff_minus_conect_perc], align='center')
plt.xticks(range(len(cutoff_minus_conect_perc)), [i[0] for i in cutoff_minus_conect_perc])
plt.xlabel("Number of CUTOFF ligands")
plt.ylabel("%")
plt.title("CUTOFF ligand numbers")
#plt.ylim(0, 65)
plt.autoscale()
plt.savefig('CUTOFF_ligand_numbers.png')
```
