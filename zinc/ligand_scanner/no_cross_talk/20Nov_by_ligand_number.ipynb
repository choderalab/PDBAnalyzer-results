```python
results = open('20Nov_ligand_scanner_results_no_cross_talk.txt')
```

```python
results_dict = eval(results.read())
```

```python
results_dict
```

```python
conect_atoms = results_dict['CONECT_atoms']
```

```python
cutoff_atoms = results_dict['CUTOFF_atoms']
```

```python
ligand_numbers = results_dict['ligand_numbers']
```

# Start by looking at 4-coordinate in CUTOFF

```python
four_ligands_dict = {}
four_conect_ligand_numbers = []
```

```python
for i in ligand_numbers:
    if i[3] == 4:
        four_conect_ligand_numbers.append(i[2])
```

```python
from collections import Counter
```

```python
Counter(four_conect_ligand_numbers)
```

```python
for i in ligand_numbers:
    four_ligands_dict[(i[0], i[1])] = []
```

```python
zn_ligands_dict = {}

for i in ligand_numbers:
    zn_ligands_dict[(i[0], i[1])] = [[], []]
```

```python
zn_ligands_dict
```

```python
for j in conect_atoms:
    zn_ligands_dict[(j[0], j[1])][0].append(j[2])
```

```python
for j in cutoff_atoms:
    zn_ligands_dict[(j[0], j[1])][1].append(j[2])
```

```python
zn_ligands_dict
```

```python
ligand_numbers_dict = {}

for i in ligand_numbers:
    ligand_numbers_dict[(i[0], i[1])] = []
```

```python
for i in ligand_numbers:
    ligand_numbers_dict[(i[0], i[1])].append(i[2])
    ligand_numbers_dict[(i[0], i[1])].append(i[3])
```

```python
ligand_numbers_dict
```

```python
for i in ligand_numbers_dict:
    if ligand_numbers_dict[i][0] != len(zn_ligands_dict[i][0]) or ligand_numbers_dict[i][1] != len(zn_ligands_dict[i][1]):
        print(i)
```

```python
four_ligands_dict = {}

for i in zn_ligands_dict:
    if len(zn_ligands_dict[i][1]) == 4:
        four_ligands_dict[i] = zn_ligands_dict[i]
```

```python
len(four_ligands_dict)
```

```python
len(zn_ligands_dict)
```

```python
from __future__ import division
len(four_ligands_dict) / len(zn_ligands_dict)
```

```python
from collections import Counter
```

```python
four_conect_cutoff_same_count = 0

for i in four_ligands_dict:
    if Counter(four_ligands_dict[i][0]) == Counter(four_ligands_dict[i][1]):
        four_conect_cutoff_same_count += 1
        
four_conect_cutoff_same_count        
        
       
```

```python
four_ligands_dict_the_unequal_ones = {}

for i in four_ligands_dict:
    if Counter(four_ligands_dict[i][0]) != Counter(four_ligands_dict[i][1]):
        four_ligands_dict_the_unequal_ones[i] = four_ligands_dict[i]
```

```python
four_ligands_dict_the_unequal_ones
```

```python
for i in four_ligands_dict_the_unequal_ones:
    four_ligands_dict_the_unequal_ones[i].append(Counter(four_ligands_dict_the_unequal_ones[i][0]) - Counter(four_ligands_dict_the_unequal_ones[i][1]))
    four_ligands_dict_the_unequal_ones[i].append(Counter(four_ligands_dict_the_unequal_ones[i][1]) - Counter(four_ligands_dict_the_unequal_ones[i][0]))
```
